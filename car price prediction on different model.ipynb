```python
import pandas as pd
import numpy as np
```

```python
car=pd.read_csv('Cleaned Car.csv')
```

```python
X=car.drop(columns='Price')
y=car['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn import tree
from sklearn.ensemble import RandomForestRegressor
import xgboost as xgb
clf1=tree.DecisionTreeRegressor()
clf2=RandomForestRegressor()
clf3=xgb.XGBRegressor()
```

```python
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe=OneHotEncoder()
ohe.fit(X[['name','company','fuel_type']])
```

```python
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),remainder='passthrough')
```

```python
pipe1=make_pipeline(column_trans,clf1)
pipe2=make_pipeline(column_trans,clf2)
pipe3=make_pipeline(column_trans,clf3)
```

```python
pipe1.fit(X_train,y_train)
pipe2.fit(X_train,y_train)
pipe3.fit(X_train,y_train)
```

```python
y_pred1=pipe1.predict(X_test)
y_pred2=pipe2.predict(X_test)
y_pred3=pipe3.predict(X_test)
```

```python
print(r2_score(y_test,y_pred1))
print(r2_score(y_test,y_pred2))
print(r2_score(y_test,y_pred3))
```

```python
scores=[]
for i in range(1000):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=i)
    clf1=tree.DecisionTreeRegressor()
    pipe1=make_pipeline(column_trans,clf1)
    pipe1.fit(X_train,y_train)
    y_pred1=pipe1.predict(X_test)
    scores.append(r2_score(y_test,y_pred1))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=np.argmax(scores))
clf1=tree.DecisionTreeRegressor()
pipe1=make_pipeline(column_trans,clf1)
pipe1.fit(X_train,y_train)
y_pred1=pipe1.predict(X_test)
r2_score(y_test,y_pred1)
```
